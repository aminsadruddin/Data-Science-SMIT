1. An employee has:


Monthly salary = 50,000
Bonus = 10,000
Deduction = 5,000

Calculate the final salary after adding bonus and subtracting deduction.


```python
monthly_salary = 50000
bonus = 10000
deduction = 50000
final_salary = monthly_salary + bonus - deduction
print(f"The Final Salary is {final_salary}")
```

2. A student obtained marks in 3 subjects:


Math = 85
Physics = 78
Computer = 92

Calculate:
Total marks
Average marks



```python
math_marks = 85
physics_marks = 78
computer_marks = 92
total_marks = math_marks + physics_marks + computer_marks
average_marks = total_marks / 3
print(f"The Average Marks is {average_marks}")
```


3. A car travels:
Speed = 60 km/h
Time = 2.5 hours


Calculate the total distance traveled.
(Hint: Distance = Speed × Time)



```python
speed = 60
time = 2.5
distance = speed * time
print(f"The Total Distance is {distance}")
```

5. A factory produces 125 items.
Each box can hold 12 items.
Calculate:

Number of full boxes
Remaining items
(Hint: Use // and %)

```python
items_produced = 125
item_per_box = 12
full_boxes = items_produced // item_per_box
remaining_items = items_produced % item_per_box # modules
print(f"The Number of Full Boxes is {full_boxes}")
print(f"The Remaining Items is {remaining_items}  -- eg of modules")
```
