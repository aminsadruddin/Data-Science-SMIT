```python

# =====================================
# 🎬 AI MOVIE RECOMMENDATION SYSTEM
# =====================================

print("===================================")
print("🎬 Welcome to AI Movie System 🎬")
print("===================================\n")


# -----------------------------
# MOVIE DATASET
# -----------------------------
movies = [
    ("Inception        ", "Sci-Fi", 9.0, 13),
    ("Interstellar     ", "Sci-Fi", 8.6, 13),
    ("The Dark Knight  ", "Action", 9.0, 13),
    ("Avengers: Endgame", "Action", 8.4, 13),
    ("Titanic          ", "Romance", 7.9, 13),
    ("The Notebook     ", "Romance", 7.8, 13),
    ("Joker            ", "Drama", 8.4, 18),
    ("Forrest Gump     ", "Drama", 8.8, 13),
    ("Toy Story        ", "Animation", 8.3, 0),
    ("Frozen           ", "Animation", 7.4, 0)
]

valid_genres = ["Action", "Sci-Fi", "Romance", "Drama", "Animation"]


# =============================
# MAIN LOOP
# =============================
while True:

    # -------------------------
    # USER NAME
    # -------------------------
    while True:
        user_name = input("Enter your name: ")

        if user_name.strip() == "":
            print("❌ Name cannot be empty")
            print("❌ Please enter a valid name.")
        # Number check
        elif user_name.isdigit():
            print("❌ Numbers are not allowed in name")
            print("❌ Please enter a valid name.")
        else:
            break



    # -------------------------
    # USER AGE
    # -------------------------
    while True:
        age_input = input("Enter your age: ")

        if age_input.isdigit():
            user_age = int(age_input)
            break
        else:
            print("❌ Age must be a number")


    # -------------------------
    # GENRE MENU
    # -------------------------
    print("\nSelect your preferred genre:")

    print("1. Action")
    print("2. Sci-Fi")
    print("3. Romance")
    print("4. Drama")
    print("5. Animation")
    print("6. All Genres") #####

    while True:
        choice = input("Enter choice (1-6): ")

        if choice.isdigit():
            choice = int(choice)

            if choice == 6:
                preferred_genre = "All"
                break

            elif 1 <= choice <= 5:
                preferred_genre = valid_genres[choice - 1]
                break

            else:
                print("❌ Choose between 0 and 5")
        else:
            print("❌ Please enter a number")


    # -------------------------
    # OPTIONAL MOOD
    # -------------------------
    user_mood = input("\nEnter mood (optional): ")

    if user_mood.strip() == "":
        user_mood = "Not provided"


    # -------------------------
    # USER PROFILE
    # -------------------------
    print("\n===================================")
    print("👤 USER PROFILE")
    print("===================================")
    print("Name:", user_name)
    print("Age:", user_age)
    print("Preferred Genre:", preferred_genre)
    print("Mood:", user_mood)


    # -------------------------
    # RECOMMENDATION LOGIC
    # -------------------------
    recommended = []

    for movie in movies:
        name, genre, rating, age_limit = movie

        # AGE CHECK
        if user_age >= 18:
            age_ok = True
        else:
            age_ok = user_age >= age_limit

        # GENRE CHECK
        if preferred_genre == "All":
            genre_ok = True
        else:
            genre_ok = (genre == preferred_genre)

        # FINAL CHECK
        if age_ok and genre_ok:
            recommended.append(movie)

    # Sort by rating (high → low)
    recommended.sort(key=lambda x: x[2], reverse=True)


    # -------------------------
    # DISPLAY RESULTS
    # -------------------------
    print("\n======================================================================================================")
    print(f"                                    🎬 Recommended Movies for {user_name}")
    print("======================================================================================================")

    if len(recommended) == 0:
        print("❌ No movies found for your selection.")
    else:
        for movie in recommended:
            name, genre, rating, age_limit = movie
            print(f"⭐ {name} \t\t\t| {genre} \t| Rating: {rating} \t| Age: {age_limit}+")


    # -------------------------
    # RESTART SYSTEM
    # -------------------------
    again = input("\nDo you want another recommendation? (yes/no): ").lower()

    if again != "yes":
        print("\n👋 Thank you for using AI Movie Recommendation System!")
        break
```
